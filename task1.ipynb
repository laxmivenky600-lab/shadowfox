5.For Loop

```python
import random
count6=0;
count1=0;
two6rows=0;
previous=0;
for i in range(20):
    roll=random.randint(1,6)
    print("roll",i+1,":",roll)
    if roll==6:
        count6+=1
    if roll==1:
        count1+=1
    if roll==6 and previous==6:
        two6rows+=1
    previous=roll
print("number of times rolled 6:",count6)
print("Number of items rolled 1:",count1)
print("Two 6s in a row:",two6rows)
```

7.File Handling

```python
students = {}

with open("/content/student_marks.xls", "r") as file:
    lines = file.readlines()

for line in lines[1:]:   # skip header
    data = line.strip().split(",")

    name = data[0]

    marks = []
    for m in data[1:]:
        if m.isdigit():        # only convert numbers
            marks.append(int(m))

    total = sum(marks)
    average = total / len(marks)

    students[name] = {
        "marks": marks,
        "total_marks": total,
        "average": average
    }

with open("new_student_marks.csv", "w") as newfile:
    newfile.write("Name,Total,Average\n")

    for student in students:
        newfile.write(
            student + "," +
            str(students[student]["total_marks"]) + "," +
            str(students[student]["average"]) + "\n"
        )
```

```python
class Avenger:

    def __init__(self, name, age, gender, power, weapon):
        self.name = name
        self.age = age
        self.gender = gender
        self.power = power
        self.weapon = weapon

    def show_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Gender:", self.gender)
        print("Super Power:", self.power)
        print("Weapon:", self.weapon)

    def is_leader(self):
        if self.name == "Captain America":
            return True
        else:
            return False


hero1 = Avenger("Captain America",100,"Male","Super Strength","Shield")
hero2 = Avenger("Iron Man",48,"Male","Technology","Armor")
hero3 = Avenger("Black Widow",35,"Female","Superhuman","Batons")
hero4 = Avenger("Hulk",45,"Male","Unlimited Strength","No Weapon")
hero5 = Avenger("Thor",1500,"Male","Super Energy","Mjolnir")
hero6 = Avenger("Hawkeye",40,"Male","Fighting Skills","Bow and Arrows")

hero1.show_info()
print("Is Leader:", hero1.is_leader())
```

9.inheritence

```python
class MobilePhone:

    def __init__(self, screenType, networkType, dualSim, frontCamera, rearCamera, ram, storage):
        self.screenType = screenType
        self.networkType = networkType
        self.dualSim = dualSim
        self.frontCamera = frontCamera
        self.rearCamera = rearCamera
        self.ram = ram
        self.storage = storage


    def make_call(self, number):
        print("Calling", number)

    def receive_call(self):
        print("Receiving a call")

    def take_picture(self):
        print("Picture taken using", self.rearCamera, "rear camera")

    def show_details(self):
        print("Screen Type:", self.screenType)
        print("Network Type:", self.networkType)
        print("Dual SIM:", self.dualSim)
        print("Front Camera:", self.frontCamera)
        print("Rear Camera:", self.rearCamera)
        print("RAM:", self.ram)
        print("Storage:", self.storage)
        print()



class Apple(MobilePhone):

    def __init__(self, screenType, networkType, dualSim, frontCamera, rearCamera, ram, storage):
        super().__init__(screenType, networkType, dualSim, frontCamera, rearCamera, ram, storage)
        self.brand = "Apple"



class Samsung(MobilePhone):

    def __init__(self, screenType, networkType, dualSim, frontCamera, rearCamera, ram, storage):
        super().__init__(screenType, networkType, dualSim, frontCamera, rearCamera, ram, storage)
        self.brand = "Samsung"



iphone1 = Apple("Touch Screen", "5G", False, "12MP", "48MP", "4GB", "64GB")
iphone2 = Apple("Touch Screen", "5G", False, "16MP", "48MP", "4GB", "64GB")


samsung1 = Samsung("Touch Screen", "4G", True, "8MP", "32MP", "3GB", "32GB")
samsung2 = Samsung("Touch Screen", "5G", True, "16MP", "48MP", "4GB", "64GB")

iphone1.show_details()
iphone1.make_call("98765***80")
iphone1.take_picture()

print()

samsung1.show_details()
samsung1.receive_call()
samsung1.take_picture()
```
